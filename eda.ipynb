```python
import matplotlib.pyplot as plt
import os
from PIL import Image # Pillow library for image processing
import numpy as np   # For creating blank placeholder images

# --- IMPORTANT: The base path where your game folders are located ---
# As per your request, this is 'outputs'.
base_path = 'outputs'

# List to store the names of the game folders found
game_folder_names = []

print(f"Current working directory: {os.getcwd()}")
print(f"Checking for game folders under: {os.path.abspath(base_path)}")

# Dynamically get all directory names under the base_path
if os.path.exists(base_path) and os.path.isdir(base_path):
    for entry_name in os.listdir(base_path):
        full_path = os.path.join(base_path, entry_name)
        if os.path.isdir(full_path):
            game_folder_names.append(entry_name)
    game_folder_names.sort() # Sort the folder names for consistent display order
else:
    print(f"ERROR: The base path '{os.path.abspath(base_path)}' does not exist or is not a directory.")
    print("Please ensure you have a folder named 'outputs' in your current working directory,")
    print("and that it contains subfolders for each game (e.g., 'Deadlock_60M_2025-05-06').")
    print("Example expected structure:")
    print(f"{os.getcwd()}/")
    print(f"└── outputs/")
    print(f"    ├── Deadlock_60M_2025-04-27/")
    print(f"    │   └── image/")
    print(f"    │       └── categorical_pie_charts.png")
    print(f"    └── Deadlock_60M_2025-05-06/")
    print(f"        └── image/")
    print(f"            └── categorical_pie_charts.png")
    # Exit or handle gracefully if the base_path is not valid
    exit()

if not game_folder_names:
    print(f"No game folders found inside '{os.path.abspath(base_path)}'.")
    print("Please ensure folders like 'Deadlock_60M_YYYY-MM-DD' exist directly inside your 'outputs' folder.")
    exit()

# Determine the grid layout for displaying images
num_images = len(game_folder_names)
cols = 3  # Number of columns in the display grid (you can adjust this)
rows = (num_images + cols - 1) // cols # Calculate the number of rows needed

plt.figure(figsize=(cols * 5, rows * 4)) # Adjust overall figure size for better visibility

print(f"Found {num_images} game folders in '{os.path.abspath(base_path)}'. Displaying 'categorical_pie_charts.png' from each.")

for i, folder_name in enumerate(game_folder_names):
    # Construct the full path to the image file
    image_path = os.path.join(base_path, folder_name, 'images', 'categorical_pie_charts.png')

    plt.subplot(rows, cols, i + 1) # Select the current subplot
    plt.title(folder_name)         # Set title for the subplot (the folder name itself)

    if os.path.exists(image_path):
        try:
            img = Image.open(image_path) # Load the image using Pillow
            plt.imshow(img)              # Display the image
            plt.axis('off')              # Hide axes for a cleaner look
        except Exception as e:
            # Handle cases where image file exists but is corrupted/unreadable
            print(f"Error loading image '{image_path}': {e}")
            plt.imshow(np.zeros((100, 100, 3), dtype=np.uint8)) # Display a black placeholder
            plt.text(0.5, 0.5, 'Error loading image',
                     horizontalalignment='center', verticalalignment='center',
                     transform=plt.gca().transAxes, color='red', fontsize=10)
            plt.axis('off')
    else:
        # Handle cases where the image file does not exist
        print(f"Image not found: '{image_path}'")
        plt.imshow(np.zeros((100, 100, 3), dtype=np.uint8)) # Display a black placeholder
        plt.text(0.5, 0.5, 'Image not found',
                 horizontalalignment='center', verticalalignment='center',
                 transform=plt.gca().transAxes, color='gray', fontsize=10)
            # Display a more detailed path for debugging
        plt.text(0.5, 0.1, f"Expected: {os.path.abspath(image_path)}",
                 horizontalalignment='center', verticalalignment='center',
                 transform=plt.gca().transAxes, color='red', fontsize=8)
        plt.axis('off')

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()         # Display the combined plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define the dataset path and files
dataset_path = '/mnt/shared/datasets/game-traces/processed_data/'
files = [
    "cyberpunk_2077_combined.parquet",
    "hogwarts_legacy_combined.parquet",
    "marvel_rivals_60m_2025-05-06_combined.parquet",
    "overwatch_60m_2025-04-28_combined.parquet",
    "the_finals_30m_2025-05-01_combined.parquet",
    "the_last_of_us_part_i.parquet",
    "combined_deadlock.parquet"
]

# Numerical columns to plot
numerical_columns = ['IO Time', 'Disk SrvT', 'IO Size', 'Duration_s', 'Throughput_MB_s', 'QD/C']

# Set style for better visuals
sns.set_style("whitegrid")

# Process each dataset individually
for file in files:
    file_path = os.path.join(dataset_path, file)
    if os.path.exists(file_path):
        # Read the Parquet file
        df = pd.read_parquet(file_path)
        
        # --- Data Processing ---
        # Clean and standardize IO Type
        df['IO Type'] = df['IO Type'].str.strip().str.lower().str.capitalize()

        # Convert hexadecimal columns to integers
        df['IO Size'] = df['IO Size'].apply(lambda x: int(x, 16))
        df['Byte Offset'] = df['Byte Offset'].apply(lambda x: int(x, 16))

        # Extract Process Name and PID
        df[['Process Name', 'PID']] = df['Process Name ( PID)'].str.extract(r'(.+)\s*\(\s*(\d+)\)')
        df['PID'] = pd.to_numeric(df['PID'], errors='coerce')

        # --- Feature Engineering ---
        # f.write("Performing feature engineering...\n")
        df['Duration_s'] = df['IO Time'] / 1_000_000_000  # Convert nanoseconds to seconds
        df['Throughput_MB_s'] = (df['IO Size'] / (1024 * 1024)) / (df['Duration_s'])
        df.loc[(df['Duration_s'] == 0) | (df['Throughput_MB_s'] == np.inf), 'Throughput_MB_s'] = 0
        
        # Create a figure with subplots
        n_cols = 2  # Number of columns in the grid
        n_rows = (len(numerical_columns) + 1) // n_cols  # Calculate rows needed
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 4 * n_rows), sharex=False, sharey=False)
        axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration
        
        # Plot distribution for each column
        for idx, column in enumerate(numerical_columns):
            if column in df.columns:
                # Use log scale for skewed distributions (e.g., IO Time, IO Size)
                if column in ['IO Time', 'Disk SrvT', 'IO Size', 'Duration_s']:
                    sns.histplot(data=df, x=column, ax=axes[idx], kde=True, bins=30, log_scale=True)
                else:
                    sns.histplot(data=df, x=column, ax=axes[idx], kde=True, bins=30)
                
                axes[idx].set_title(f'{file} - {column} Distribution')
                axes[idx].set_xlabel(column)
                axes[idx].set_ylabel('Count')
                
                # Set consistent x-axis limits based on data range
                if column in df.columns:
                    x_min, x_max = int(df[column].min()), int(df[column].max())
                    if x_min > 0 and x_max > 0:  # Avoid log scale issues with zero or negative values
                        axes[idx].set_xlim(left=max(1, x_min * 0.9), right=x_max * 1.1)
            else:
                axes[idx].text(0.5, 0.5, f'{column} not found', ha='center', va='center', transform=axes[idx].transAxes)
        
        # Remove empty subplots if any
        for idx in range(len(numerical_columns), len(axes)):
            fig.delaxes(axes[idx])
        
        # Adjust layout to prevent overlap
        plt.tight_layout()
        
        # Save the plot with a descriptive name
        output_file = os.path.join(os.getcwd(), f'distribution_{os.path.splitext(file)[0]}.png')
        plt.savefig(output_file)
        print(f"Saved distribution plot for {file} to: {output_file}")
        
        # Display the plot
        plt.show()
        
        # Close the figure to free memory
        plt.close(fig)
    else:
        print(f"Warning: File not found - {file_path}")

print("All distribution plots have been generated and displayed.")
```
